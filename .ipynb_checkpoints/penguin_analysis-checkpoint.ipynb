```python
import pandas as pd
penguins_df = pd.read_csv("penguins_mass.csv")
penguins_df
```

```python
from Bio import SeqIO
from Bio.SeqUtils import molecular_weight
import matplotlib.pyplot as plt
```

```python
records = list(SeqIO.parse("penguins_cytb.fasta", "fasta"))
len(records)
```

```python
print(records[0].id)
print(records[0].seq[:60])
```

```python
records[0].seq
```

```python
def get_gc_content(dna_seq):
    dna_seq = str(dna_seq).upper()
    g_count = dna_seq.count("G")
    c_count = dna_seq.count("C")
    gc_content = (g_count + c_count) / len(dna_seq) * 100
    return gc_content
```

```python
get_gc_content(records[0].seq)
```

```python
def translate_dna(dna_seq):
    return dna_seq.translate()
```

```python
aa_seq = translate_dna(records[0].seq)
aa_seq
```

```python
def get_molecular_weight(aa_seq):
    aa_seq = str(aa_seq).replace("*", "")
    return molecular_weight(aa_seq, seq_type="protein")
```

```python
get_molecular_weight(aa_seq)
```

```python
sequence_data = {}

for record in records:
    species = record.id
    dna_seq = record.seq
    aa_seq = translate_dna(dna_seq)
    gc = get_gc_content(dna_seq)
    mw = get_molecular_weight(aa_seq)
    
    sequence_data[species] = {
        "GC_content": gc,
        "molecular_weight": mw
    }

sequence_data
```

```python
seq_df = pd.DataFrame.from_dict(sequence_data, orient="index")
seq_df.reset_index(inplace=True)
seq_df.rename(columns={"index": "species"}, inplace=True)
seq_df
```

```python
sequence_data = {}

for record in records:
    description_parts = record.description.split()
    species = description_parts[1] + " " + description_parts[2]
    
    dna_seq = record.seq
    aa_seq = translate_dna(dna_seq)
    gc = get_gc_content(dna_seq)
    mw = get_molecular_weight(aa_seq)
    
    sequence_data[species] = {
        "GC_content": gc,
        "molecular_weight": mw
    }
```

```python
seq_df = pd.DataFrame.from_dict(sequence_data, orient="index")
seq_df.reset_index(inplace=True)
seq_df.rename(columns={"index": "species"}, inplace=True)
seq_df
```

```python
final_df = pd.merge(penguins_df, seq_df, on="species")
final_df
```

```python
plt.scatter(final_df["GC_content"], final_df["mass"])
plt.xlabel("GC Content (%)")
plt.ylabel("Body Mass (kg)")
plt.title("Penguin Body Mass vs GC Content")
plt.show()
```

```python
plt.scatter(final_df["molecular_weight"], final_df["mass"])
plt.xlabel("Molecular Weight")
plt.ylabel("Body Mass (kg)")
plt.title("Penguin Body Mass vs Molecular Weight")
plt.show()
```
